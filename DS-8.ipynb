```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Fare'].fillna(df['Fare'].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
sbn.countplot(x='Sex', data=df)
```

```python
sbn.barplot(data=df, x='Sex', y='Fare')
```

```python
sbn.histplot(data=df, x='Age')
```

```python
sbn.jointplot(data=df, x='Age', y='Fare')
```

```python
sbn.jointplot(data=df, x='Age', y='Fare', kind='hex')
```

```python
sbn.boxplot(data=df, x='Sex', y='Age')
```

```python
sbn.boxplot(data=df, x='Sex', y='Age', hue='Survived')
```

```python
sbn.violinplot(data=df, x='Sex', y='Age')
```

```python
sbn.violinplot(data=df, x='Sex', y='Age', hue='Survived')
```

```python
sbn.stripplot(data=df, x='Sex', y='Age')
```

```python
sbn.stripplot(data=df, x='Sex', y='Age', jitter=False)
```

```python
sbn.swarmplot(data=df, x='Sex', y='Age')
```

```python
sbn.swarmplot(data=df, x='Sex', y='Age',hue='Survived')
```

```python
sbn.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
```
